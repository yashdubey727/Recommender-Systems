```python
## Aim: Recommend Movies similar to the movies,'Liar Liar' and 'Star Wars' using the kaggle dataset on Movies.
```

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']

#read the data

movie_data = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
movie_data.head()
```

```python
titles = pd.read_csv("Movie_Id_Titles")
titles.head()
```

```python
#merge them on item_id
movie_data = pd.merge(movie_data,titles,on='item_id')
movie_data.head()
```

```python
# create a ratings dataframe with average rating and number of ratings

movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
#create a ratings data frame
rating_df = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
rating_df.head()
```

```python
rating_df['number of ratings'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
rating_df.head()
```

```python
plt.figure(figsize=(10,4))
rating_df['number of ratings'].plot(kind='kde')
```

Now let's create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies.

```python
movie_watch = movie_data.pivot_table(index='user_id',columns='title',values='rating')
movie_watch.head()
```

```python
rating_df.sort_values('number of ratings',ascending=False).head(10)
```

```python
#take example of two movies = 'liar liar' and 'starwars'
#recommend movies similar to them
starwars_user_ratings = movie_watch['Star Wars (1977)']
liarliar_user_ratings = movie_watch['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
#use corrwith() 
similar_to_starwars = movie_watch.corrwith(starwars_user_ratings)
similar_to_liarliar = movie_watch.corrwith(liarliar_user_ratings)
```

```python
#remove the missing values from the dataset
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
#there is some problem with the result
# some users have watched only one movie and gave it a 5 star rating, but this can lead to inaproropriate results
```

```python
corr_starwars = corr_starwars.join(rating_df['number of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
# the problem is solved when we sort values on number of ratings
```
